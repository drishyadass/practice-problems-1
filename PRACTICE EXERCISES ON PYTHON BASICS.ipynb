```python
#Write a Python program to check whether a number is prime or not.
#Sample input: Enter the number: 3
#Sample output: 3 is a prime number


n = int(input("enter a numb:  "))

if n>1:
    for i in range(2, n):
      if n%i == 0:
        print(n,"is not a primer number")
        break
    else:
      print(n, "is a prime number")
else:
    print(n, "is not a prime number")
```

```python
#Write a Python program to reverse a list using for loop.

n = [10,40,30,70]

for i in range(len(n)):
  reverse = n.pop()
  n.insert(i,reverse)
  print(n)
```

```python
#Write a Python program to construct the following pattern, using a
#nested for loop

n = 5

for i in range(n):
    for j in range(i):
        print('* ', end="")
    print('')

for i in range(n, 0, -1):
    for j in range(i):
        print('* ', end="")
    print('')
```
